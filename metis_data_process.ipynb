```python
import pandas as pd
```

```python
import sqlalchemy as db
from sqlalchemy.engine import create_engine
from urllib.parse import quote
import psycopg2

def read_metis_dist():
    engine = create_engine('postgresql://metis:%s@192.168.2.102/metisdb' % quote('password'))
    connection = engine.connect()
    df_resource = pd.read_sql("select * from resource", engine)
    return df_resource

# 1. metis 192.168.2.102 메티스 디비에 접속하는 엔진을 만든다.
# 2. 엔진을 연결해주는 connection을 만들어준다. 데이터를 쏴줄때는 필요하지만, 현재는 읽기만 하는 거라 필요가 없다.
# 3. df_resource안에 판다스 sql 메소드로 엔진안에 있는 sql을 읽어준다. (select * from resource - 데이터베이스 안에 있는 resource 테이블만 읽어준다.)
# 4. return df_resource
```

```python
df_resource = read_metis_dist()
before_title = list(df_resource['title'])
```

```python
df_present = pd.read_csv('메타데이터_최종.csv')
after_title = list(df_present['title'])
```

```python
dup_list = [x for x in after_title if x in before_title]
# 만약 X가 before_title 리스트 안에 있을 때, after_title의 안에 있는 X를 하나씩 뽑음
# x는 after_title과 before_title의 교집합
```

```python
fin_df_present = pd.read_csv('메타데이터_최종.csv', index_col=0)
fin_df_present_test = fin_df_present.drop(dup_list)
```
